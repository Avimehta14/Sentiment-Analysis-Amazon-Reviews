# Amazon Data Sentiment Analysis

Importing and Downloading Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk
import io
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords


nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

Removing Lemmatization

```python
stop_words = set(stopwords.words('english'))
wordnet_lemmatizer = WordNetLemmatizer() 
```

Reading CSV and converting to Dataframe

```python
df = pd.read_csv("Data1_new")
df
```

Normalizing the Text (Tokenization, Stopword Removal, etc.)

```python
def normalizer(review):
    only_letters = re.sub("[^a-zA-Z]", " ",review) 
    tokens = nltk.word_tokenize(only_letters)[2:]    # Tokenize data by words
    lower_case = [l.lower() for l in tokens]          # Converting data into lowercase
    filtered_result = list(filter(lambda l: l not in stop_words, lower_case))   # Removing stopwords
    lemmas = [wordnet_lemmatizer.lemmatize(t) for t in filtered_result]         # Lemmatizing the data  
    
    s=" "
    s= s.join(lemmas)
    return s
```

Finding Polarity and Subjectivity of each Review

```python
analy=[]
subj=[]
pol=[]

for i in range(len(df['Reviews'])):
    tex = df.iloc[i,5]
    op = normalizer(tex)
    analy.append(op)
    anpol=TextBlob(op).polarity
    anali=TextBlob(op).subjectivity
    subj.append(anali)
    pol.append(anpol)

analy    
```

Printing the DataFrame with Additional Parameters

```python
df['Review_normalized']=analy
df['Subjectivity_Review']=subj
df['Polarity_Review']=pol
df
```

```python
h_analy=[]
h_subj=[]
h_pol=[]

for i in range(len(df['Heading'])):
    tex = df.iloc[i,2]
    op = normalizer(tex)
    h_analy.append(op)
    h_anpol=TextBlob(op).polarity
    h_anali=TextBlob(op).subjectivity
    h_subj.append(anali)
    h_pol.append(anpol)
    
h_analy
```

```python
df['heading_normalized']=h_analy
df['Subjectivity_Heading']=h_subj
df['Polarity_Heading']=h_pol
df
```

Finally converting Dataframe to CSV

```python
df.to_csv("Data_final_1_new",index=False)
```
