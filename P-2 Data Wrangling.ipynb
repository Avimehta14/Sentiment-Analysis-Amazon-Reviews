Fetching data from CSV related earlier

```python
df = pd.read_csv("Mobiles")
df
```

Cleaning the Name of the Product

```python
#cleaning the name

ext=[]
for i in range(len(df['Product'])):
    name , extra = df.iloc[i,0].split("(")
    df.iloc[i,0]=name
    ext.append(extra)
```

Cleaning Price Factor

```python
# cleaning price

pr = df.iloc[0,1]
df.iloc[0,1]= float(pr[1:])

pr2 = df.iloc[1,1]
df.iloc[1,1] = 0.0

pr3 = df.iloc[2,1]
df.iloc[2,1]= float(pr3[1:])

pr4 = df.iloc[3,1]
df.iloc[3,1]= float(pr4[1:])
```

Clean Rating and Availibility Factors

```python
#cleaning ratings 

for i in range(len(df["Ratings"])):
    val = df.iloc[i,2]
    w = val.split(" ")
    df.iloc[i,2] = float(w[0])

#clean total ratings 

for i in range(len(df["Total Ratings"])):
    val = df.iloc[i,3]
    w = list(val.split(" "))
    df.iloc[i,3] = float(w[0])

#cleaning availibilty 

df.iloc[0,4]= "Yes"
df.iloc[1,4] = "No"
df.iloc[2,4]= "Yes"
df.iloc[3,4]= "Yes"

df
```

```python
#Product Name clean

ext=[]
for i in range(len(df['Product Name'])):
    name , extra = df.iloc[i,0].split("(")
    df.iloc[i,0] = name
    ext.append(extra)
```

```python
# Dates clean

for i in range(len(df["Date"])):
    rw= df.iloc[i,3]
    temp = [ x for x in rw.split(" ")]
    tem  = str(temp[4:])
    st=" ".join(temp[4:])
    #print(st)
    df.iloc[i,3] = st

# Cleaning Ratings

for i in range(len(df["Ratings"])):
    rat = df.iloc[i, 4]
    df.iloc[i,4]= float(rat[0:3])
df    
```

Converting to CSV file
